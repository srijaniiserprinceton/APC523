# 4. Conditioning

**(c)** We found that the (condA) ~ $\frac{e^{x} - 1}{x}$. Now we know that smaller the condition number, the better is the algorithm posed. So, we try to set an upper bound for the condition number in the domain $x ~ [0,1]$. In doing so, we maximize the numerator which gives us (condA) <= $\frac{e - 1}{x}$. However, the x in the denominator makes our expression larger as we approach x = 0.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
plt.ion()
condA = lambda x: ((np.exp(1) -1)/x)  #defining our condition of A 
condf = lambda x: x/(np.exp(x) -1)  #defining the condition of f
x = np.arange(1e-1,1,0.001)
```

```python
plt.plot(x,condA(x),label='condA(x)')
plt.plot(x,condf(x), label ='condf(x)')
plt.plot(x,np.ones(len(x)),'--k',label='y = 1.0')
plt.xlabel('x')
plt.legend()
```
